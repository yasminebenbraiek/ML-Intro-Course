```python
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
path_to_png_file = "nb.jpg"
import matplotlib.image as mpimg
img = mpimg.imread(path_to_png_file) #from image to pixel matrix
print(img.shape) #lignes, colonnes, rgb; les dimensions de 199 et 234 se repetent 3 fois en superposition
print(img) #les exemples du modele sont des pixels, caracteristiques 3 rgb
```

```python
pixels = [pixel for row in img for pixel in row]
print(np.array(pixels).shape)
clusterer = KMeans(n_clusters=4, random_state=0).fit(pixels)
# for i in clusterer.labels_:
#  print(i)
print(clusterer.labels_) #etiquette chaque exemple, dans ce cas les pixels
print(clusterer.cluster_centers_) #moyenne de chaque groupe
```

```python
def recolor(pixel):
  cluster = clusterer.predict(np.array([pixel],int))
  return clusterer.cluster_centers_[cluster][0]
```

```python
new_img = [[recolor(pixel) for pixel in row] for row in img]
new_img=np.array(new_img,int)
print(new_img.shape)
plt.imshow(new_img)
plt.show() #from pixels to image
```
